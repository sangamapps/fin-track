```python
import pandas as pd
import json


def get_data(filepath):
    df = pd.read_table(
        filepath,
        # encoding="utf-8",
        # delimiter="\t",
        on_bad_lines="warn",
        names=[
            "Txn Date",
            "Value Date",
            "Description",
            "Ref No./Cheque No.",
            "Debit",
            "Credit",
            "Balance",
        ],
        skip_blank_lines=True,
    )
    df = df.dropna(how="all")
    df = df.fillna("")
    return df.values.tolist()


data = get_data("/Users/asangame/Downloads/SBI 01-01-2025 to 28-02-2025.xls")
print(len(data))
print(json.dumps(data, indent=4))
```

```python
import pandas as pd
import json


def get_data(filepath):
    df = pd.read_excel(
        filepath,
        engine="xlrd",
        names=[
            "Date",
            "Narration",
            "Chq./Ref.No.",
            "Value Dt",
            "Withdrawal Amt.",
            "Deposit Amt.",
            "Closing Balance",
        ],
    )
    df = df.dropna(how="all")
    df = df.fillna("")
    return df.values.tolist()


print(
    json.dumps(
        get_data("/Users/asangame/Downloads/HDFC 01-01-2025 to 28-02-2025.xls"), indent=4
    )
)
```
